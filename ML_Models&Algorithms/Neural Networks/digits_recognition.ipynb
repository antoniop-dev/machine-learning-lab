```python
import pandas as pd
from ucimlrepo import fetch_ucirepo 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, Perceptron
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, confusion_matrix

RANDOM_SEED = 0
```

```python
# fetch dataset 
optical_recognition_of_handwritten_digits = fetch_ucirepo(id=80) 
  
# data (as pandas dataframes) 
X = optical_recognition_of_handwritten_digits.data.features 
y = optical_recognition_of_handwritten_digits.data.targets 
  
# metadata 
print(optical_recognition_of_handwritten_digits.metadata) 
  
# variable information 
print(optical_recognition_of_handwritten_digits.variables) 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=RANDOM_SEED)

X_max = max(X_train.max())
X_train /= X_max
X_test /= X_max
```

### Logistic Regression

```python
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train.values.ravel())
print("Logistic Regression report:")
print(classification_report(y_test, lr.predict(X_test)))
print(confusion_matrix(y_test, lr.predict(X_test)))
```

### Neural Network

### Perceptron

```python
p = Perceptron()
p.fit(X_train, y_train.values.ravel())
print("Perceptron report:")
print(classification_report(y_test, p.predict(X_test)))
print(confusion_matrix(y_test, p.predict(X_test)))
```

### One Hidden Layer Network (MLP)

```python
layer_size = (100,)
one_nn = MLPClassifier(hidden_layer_sizes=layer_size, max_iter=1000, early_stopping=True)
one_nn.fit(X_train, y_train.values.ravel())
print("One Hidden Layer Netowrk (MLP) report:")
print(classification_report(y_test, one_nn.predict(X_test)))
print(confusion_matrix(y_test, one_nn.predict(X_test)))
```

### Two Hidden Layer Network

```python
layer_size = (100,100)
two_nn = MLPClassifier(hidden_layer_sizes=layer_size, max_iter=1000, early_stopping=True)
two_nn.fit(X_train, y_train.values.ravel())
print("Two Hidden Layer Netowrk (MLP) report")
print(classification_report(y_test, two_nn.predict(X_test)))
print(confusion_matrix(y_test, two_nn.predict(X_test)))
```

### Three Hidden Layers Network

```python
layer_size = (100,100,100)
three_nn = MLPClassifier(hidden_layer_sizes=layer_size, max_iter=1000, early_stopping=True)
three_nn.fit(X_train, y_train.values.ravel())
print("Three Hidden Layer Netowrk (MLP) report")
print(classification_report(y_test, three_nn.predict(X_test)))
print(confusion_matrix(y_test, three_nn.predict(X_test)))
```
